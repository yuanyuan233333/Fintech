```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, RMSprop, SGD, Adagrad
```

```python
np.random.seed(1)
f= lambda x:np.sin(x)
x_train=np.linspace(0,10,20)
x_test=np.linspace(0,10,100)+np.random.normal(0,00.1,100)
train=f(x_train)
test=f(x_test)
```

```python
# Plotting
plt.plot(x_train,train,'ro')
plt.plot(x_test,test)
```

```python
x_train=np.linspace(0,10,2000)
x_test=np.linspace(0,10,100)+np.random.normal(0,00.1,100)
train=f(x_train)
test=f(x_test)
```

```python
import tensorflow as tf 
tf.random.set_seed(1) #for reproducibility of ML results
```

```python
model=Sequential()
model.add(Dense(units=10, activation='tanh', input_dim=1))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=1, activation='linear'))#output layer
model.compile(loss='mse',
              optimizer=Adam(lr=0.001,beta_1=0.9,beta_2=0.999))

model.fit(x_train,train,epochs=2000,verbose=0,batch_size=200)
#batch_size: Integer or None. Number of samples per gradient update. 
#epochs: Integer. Number of epochs to train the model. An epoch is an iteration over the entire x and y data provided. 
#verbose: 0, 1, or 2. Verbosity mode. 0 = silent, 1 = progress bar, 2 = one line per epoch. 
    
pred_train=model.predict(x_train)
pred_test=model.predict(x_test)

print(np.sum((np.ravel(pred_train)-train)**2)/pred_train.shape[0])
print(np.sum((np.ravel(pred_test)-test)**2)/pred_test.shape[0])

plt.plot(x_train,pred_train)
```
